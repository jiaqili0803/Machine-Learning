Your answer for 1(a) here:
* the clusters we find in these two cases will not be the same, as we see the plots from lecture and lab, they are different distance methods so will have different results.
* Euclidean Distance: It measures the straight line distance between 2 points in space.
* Cosine Similarity: It measures the angular cosine similarity between 2 vectors.

Your answer for 1(b) in the following code cell:

```python
# Code for 1(b)

def answer_one():

    X = np.array([[0.5, 1.5], [2, 3], [3, 2], [1.5, 0.5]])

    cls_1 = AgglomerativeClustering(linkage = 'average', affinity = 'euclidean', n_clusters = 2)
    cluster_labels_euclidean = cls_1.fit_predict(X)
    
    cls_2 = AgglomerativeClustering(linkage = 'average', affinity = 'cosine', n_clusters = 2)
    cluster_labels_cosine = cls_2.fit_predict(X)


    return cluster_labels_euclidean, cluster_labels_cosine

answer_one()
```
